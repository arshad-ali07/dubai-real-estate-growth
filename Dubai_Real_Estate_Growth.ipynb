```python
import pandas as pd

# data for Dubai real estate growth from 2015–2025
data = {
    "Year": list(range(2015, 2026)),
    "Avg_Price_psf_AED": [1050, 1020, 980, 950, 925, 900, 1000, 1150, 1300, 1450, 1600],
    "Transaction_Value_Billion_AED": [115, 120, 110, 108, 112, 130, 150, 180, 220, 275, 325],
    "Population_Millions": [2.45, 2.55, 2.65, 2.75, 2.9, 3.1, 3.3, 3.5, 3.65, 3.8, 3.95],
    "New_Homes_Units": [25000, 26000, 27000, 28000, 29000, 30000, 32000, 34000, 36000, 39000, 41000],
    "Avg_Rent_AED_Yearly": [65000, 64000, 63000, 62000, 61000, 60000, 67000, 72000, 77000, 85000, 90300]
}


df = pd.DataFrame(data)
df.to_csv("dubai_real_estate.csv", index=False)

print("✅ File 'dubai_real_estate.csv' created successfully.")
```

```python
import pandas as pd

df = pd.read_csv("dubai_real_estate.csv")


print(df)
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Avg_Price_psf_AED', data=df, marker='o')
plt.title('Dubai Property Price Per Sq Ft (2015–2025)')
plt.xlabel('Year')
plt.ylabel('Price (AED)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Transaction_Value_Billion_AED', data=df)
plt.title('Total Real Estate Transaction Value (AED Billion)')
plt.ylabel('Transaction Value (Billion AED)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population_Millions', y='Avg_Rent_AED_Yearly', data=df)
plt.title('Population vs Average Yearly Rent')
plt.xlabel('Population (millions)')
plt.ylabel('Avg Rent (AED)')
plt.show()
```

📌 Key Insights
- Property prices increased ~70% from 2020 to 2025.
- Transaction volume more than doubled between 2015 and 2025.
- Population growth appears to correlate with rent increases.
